```python
from scipy.sparse import lil_matrix
from scipy.sparse.linalg import spsolve
from numpy.linalg import solve, norm
from numpy.random import rand
```

```python
A = lil_matrix((1000, 1000))
```

```python
A
```

```python
A[0, :100] = rand(100)
A[1, 100:200] = A[0, :100]
A.setdiag(rand(1000))
```

```python
A = A.tocsr()
b = rand(1000)
x = spsolve(A, b)
```

```python
x
```

```python
x_ = solve(A.toarray(), b)
```

```python
x_
```
